# 2. Ventas por categor√≠a

```python
#Contar productos vendidos por categor√≠a en cada tienda
ventas_por_categoria = tiendas.groupby(["Tienda", "Categor√≠a del Producto"])["Producto"].count().reset_index()
ventas_por_categoria.columns = ["Tienda", "Categor√≠a", "Cantidad Vendida"]
print("Cantidad de productos vendidos por categor√≠a en cada tienda:")
print(ventas_por_categoria)
```

# 3. Calificaci√≥n promedio de la tienda


```python
#Calcular calificaci√≥n promedio por tienda
calificaciones_promedio = tiendas.groupby("Tienda")["Calificaci√≥n"].mean().reset_index()
calificaciones_promedio.columns = ["Tienda", "Calificaci√≥n Promedio"]
print("Calificaci√≥n promedio de clientes por tienda:")
print(calificaciones_promedio)
```

# 4. Productos m√°s y menos vendidos

```python
#  Productos m√°s y menos vendidos por tienda
ventas_productos = tiendas.groupby(["Tienda", "Producto"])["Producto"].count().rename("Cantidad").reset_index()

# Obtener productos m√°s vendidos
productos_mas_vendidos = ventas_productos.loc[ventas_productos.groupby("Tienda")["Cantidad"].idxmax()].reset_index(drop=True)

# Obtener productos menos vendidos
productos_menos_vendidos = ventas_productos.loc[ventas_productos.groupby("Tienda")["Cantidad"].idxmin()].reset_index(drop=True)

print("Productos m√°s vendidos por tienda:")
print(productos_mas_vendidos)

print("\nProductos menos vendidos por tienda:")
print(productos_menos_vendidos)

# Gr√°fico de barras agrupadas: Productos vendidos por categor√≠a
plt.figure(figsize=(10, 6))
sns.barplot(data=ventas_por_categoria, x="Categor√≠a", y="Cantidad Vendida", hue="Tienda")
plt.title("Cantidad de Productos Vendidos por Categor√≠a y Tienda")
plt.ylabel("Cantidad Vendida")
plt.xlabel("Categor√≠a de Producto")
plt.xticks(rotation=45)
plt.legend(title="Tienda")
plt.tight_layout()
plt.show()
```

# **6.**  An√°lisis geogr√°fico de ventas por tienda

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tiendas, x="lon", y="lat", hue="Tienda", palette="deep", alpha=0.7)
plt.title("Distribuci√≥n Geogr√°fica de las Ventas por Tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.grid(True)
plt.legend(title="Tienda")
plt.tight_layout()
plt.show()

#mapa de calor geogr√°fico (densidad de puntos)
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=tiendas, x="lon", y="lat",
    cmap="Reds", fill=True, thresh=0.05, levels=100, alpha=0.8
)
plt.title("Mapa de Calor de Concentraci√≥n de Ventas (Todas las Tiendas)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.grid(True)
plt.tight_layout()
plt.show()

# Heatmap por tienda espec√≠fica (ejemplo con Tienda 1)
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=tiendas[tiendas["Tienda"] == "Tienda 1"], x="lon", y="lat",
    cmap="Blues", fill=True, thresh=0.05, levels=100, alpha=0.8
)
plt.title("Mapa de Calor de Ventas - Tienda 1")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.grid(True)
plt.tight_layout()
plt.show()

```
